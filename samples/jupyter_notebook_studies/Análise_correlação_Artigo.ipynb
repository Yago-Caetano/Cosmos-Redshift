```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from sklearn import metrics
```

```python
data = pd.read_csv('sample_dataset.csv',decimal=",")
data = data.tail(100) #use only the last 100 rows
```

```python
df_novo = data[['temperature', 'humidity', 'cracked']]

df_novo = df_novo.dropna(how='all',axis=0)

df_novo['humidity'] = pd.to_numeric(df_novo['humidity'], errors='coerce')
df_novo['temperature'] = pd.to_numeric(df_novo['temperature'], errors='coerce')



df_novo
```

```python
X = df_novo[['temperature','humidity']]
Y = df_novo['cracked']

X = X.reset_index(drop=True)
Y = Y.reset_index(drop=True)

valor = df_novo.iloc[0,1]

print(type(valor))
#print(type(df_novo['temperature'][0]))
```

```python
from sklearn.linear_model import LinearRegression


reg = LinearRegression().fit(X,Y)

print(f'{reg.coef_} {reg.intercept_}')

m_predict = reg.predict(X)

plt.plot(m_predict, color = 'blue')
plt.plot(Y, color = 'red')
plt.show()
```
