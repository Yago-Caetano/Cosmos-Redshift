```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from sklearn import metrics
```

```python
data = pd.read_csv('sample_dataset.csv',decimal=",")
#data = data.tail(100) #use only the last 100 rows

```

```python
df_novo = data[['temperature', 'humidity', 'cracked']]

df_novo = df_novo.dropna(how='all',axis=0)


df_novo
```

```python
from sklearn.preprocessing import StandardScaler



#Dados foram normalizados
padronizador = StandardScaler()
padronizador.fit(df_novo)
std_dt = padronizador.transform(df_novo)
std_dt = pd.DataFrame(data = std_dt,columns=df_novo.keys())


```

```python
import seaborn as sns

matriz_correlacao = std_dt.corr()
plt.figure(figsize=(5, 5)) # width and height in inches
sns.heatmap(matriz_correlacao, annot = True)
```
